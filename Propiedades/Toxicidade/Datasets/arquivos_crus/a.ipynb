```python
# Exemplo de script Python para ler um arquivo FASTA com labels e escrever sim.txt / nao.txt

# 1) Defina o caminho do seu arquivo FASTA
fasta_path = '/scratch/pedro.bacelar/HIV-Models/Propiedades/Toxicidade/Datasets/arquivos_crus/valid.fasta'

# 2) Abra os arquivos de saída
with open('sim_val.txt', 'w', encoding='utf-8') as sim_file, \
     open('nao_val.txt', 'w', encoding='utf-8') as nao_file:

    # Variáveis temporárias
    current_header = None
    seq_lines = []

    # 3) Percorra o arquivo FASTA
    with open(fasta_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            if line.startswith('>'):
                # Se já houver um registro carregado, processe-o
                if current_header is not None:
                    seq = ''.join(seq_lines)
                    parts = current_header.split()
                    try:
                        label = int(parts[-1])
                    except ValueError:
                        # pula se o label não for inteiro
                        pass
                    else:
                        if label == 1:
                            sim_file.write(seq + '\n')
                        elif label == 0:
                            nao_file.write(seq + '\n')

                # Inicie novo registro
                current_header = line
                seq_lines = []
            else:
                # Acumula linhas da sequência
                seq_lines.append(line)

        # Processa o último registro após o loop
        if current_header is not None:
            seq = ''.join(seq_lines)
            parts = current_header.split()
            try:
                label = int(parts[-1])
            except ValueError:
                pass
            else:
                if label == 1:
                    sim_file.write(seq + '\n')
                elif label == 0:
                    nao_file.write(seq + '\n')
```
